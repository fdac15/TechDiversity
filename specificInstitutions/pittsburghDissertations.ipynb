More info including URN avalable through URL

```python
import requests, json, sys, time, re
from bs4 import BeautifulSoup

mathURL = 'http://d-scholarship.pitt.edu/view/divisions/sch=5Fas=5Fmath/thesis=5Fdegree.html'
csURL = 'http://d-scholarship.pitt.edu/view/divisions/sch=5Fas=5Fcompsci/thesis=5Fdegree.html'

mathR = requests.get(mathURL)
mathT = mathR.text
csR = requests.get(csURL)
csT = csR.text

mathSoup = BeautifulSoup(mathT, 'html.parser')
csSoup = BeautifulSoup(csT, 'html.parser')

def run(soup, subject):
    entries = []
    for entry in soup.findAll('p'):
        print(entry)
        
        dictionary = {
            "title": entry.find('em').string,
            "author": entry.find('span').string,
            "institution": entry.contents[3].split(',')[1].strip(),
            "year": entry.contents[1].strip(' ()'),
            "subject": subject,
            "url": entry.find('a')['href'],
            "type": entry.contents[3].split(',')[0].strip()
        }
        print(dictionary)
        entries.append(dictionary)
    return entries

entries = run(mathSoup, 'mathematics')
entries = entries + run(csSoup, 'computer science')

with open("pittsburghDissertations.json", "w") as f:
    json.dump(entries, f, indent = 2)
```
