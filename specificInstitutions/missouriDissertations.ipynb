```python
import requests, json, sys, time, re
from bs4 import BeautifulSoup

def run(soup, subject):
    entries = []
    for entry in soup.findAll('li', {'class':'ds-artifact-item'}):
        #print(entry)
        #print('thing', entry.find('span', {'class':'author'}).contents[0].string)
        #try:
        dictionary = {
            "title": entry.find('a').string,
            "author": entry.find('span', {'class':'author'}).contents[0].string,
            "institution": 'Missouri University',
            "year": entry.find('span', {'class':'date'}).string,
            "subject": subject,
            "url": 'https://mospace.umsystem.edu' + entry.find('a')['href']
        }
        #except Exception as e:
         #   print('skipping', entry)
          #  print('error', e)
           # continue;
        if(dictionary['author'][-1] == '-'): #Some authors have some date info. Trim it
            dictionary['author'] = dictionary['author'][:-7]
        print(dictionary)
        entries.append(dictionary)
    return entries

def get_page(URL):
    r = requests.get(URL)
    t = r.text
    return BeautifulSoup(t, 'html.parser')



mathSoup = []
mathSoup.append(get_page('https://mospace.umsystem.edu/xmlui/handle/10355/3989/browse?order=ASC&rpp=100&sort_by=-1&value=Mathematics&etal=-1&offset=0&type=discipline'))
mathSoup.append(get_page('https://mospace.umsystem.edu/xmlui/handle/10355/3988/browse?order=ASC&rpp=100&sort_by=-1&value=Mathematics+%28MU%29&etal=-1&offset=0&type=discipline'))



csSoup = []
csSoup.append(get_page('https://mospace.umsystem.edu/xmlui/handle/10355/3989/browse?order=ASC&rpp=100&sort_by=-1&value=Computer+science+%28MU%29&etal=-1&offset=0&type=discipline'))
csSoup.append(get_page('https://mospace.umsystem.edu/xmlui/handle/10355/3988/browse?order=ASC&rpp=100&sort_by=-1&value=Computer+science+%28MU%29&etal=-1&offset=0&type=discipline'))

entries = []

for soup in mathSoup:
    entries = entries + run(soup, 'mathematics')
for soup in csSoup:
    entries = entries + run(soup, 'computer science')

with open("missouriDissertations.json", "w") as f:
    json.dump(entries, f, indent = 2)
```
