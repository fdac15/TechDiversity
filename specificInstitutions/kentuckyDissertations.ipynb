```python
import requests, json, sys, time, re
from bs4 import BeautifulSoup

def run(soup, subject):
    entries = []
    for entry in soup.findAll('p', {'class':'article-listing'}):
        print(entry)
        
        dictionary = {
            "title": entry.find('a').string,
            "author": entry.contents[1].strip(' ,'),
            "institution": 'University of Kentucky',
            "year": entry.find_previous('h4').contents[0].strip().split()[-1],
            "subject": subject,
            "url": entry.find('a')['href']
        }
        print(dictionary)
        entries.append(dictionary)
    return entries

def get_page(URL):
    r = requests.get(URL)
    t = r.text
    return BeautifulSoup(t, 'html.parser')

mathSoup = []
mathSoup.append(get_page('http://uknowledge.uky.edu/math_etds/'))
for i in range(1,3):
    mathSoup.append(get_page('http://uknowledge.uky.edu/math_etds/index.' + str(i) + '.html'))

csSoup = []
csSoup.append(get_page('http://uknowledge.uky.edu/cs_etds/index.html'))
for i in range(1,5):
    csSoup.append(get_page('http://uknowledge.uky.edu/cs_etds/index.' + str(i) + '.html'))

entries = []

for soup in mathSoup:
    entries = entries + run(soup, 'mathematics')
for soup in csSoup:
    entries = entries + run(soup, 'computer science')

with open("kentuckyDissertations.json", "w") as f:
    json.dump(entries, f, indent = 2)
```
