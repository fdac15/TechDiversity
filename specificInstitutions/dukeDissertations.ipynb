```python
import requests, json, sys, time, re
from bs4 import BeautifulSoup

def run(soup, subject):
    entries = []
    for entry in soup.findAll('li', {'class':'ds-artifact-item'}):
        #print(entry)
        #print('thing', entry.find('span', {'class':'author'}).contents[1].string)
        try:
            dictionary = {
                "title": entry.find('span', {'class':'Z3988'}).string,
                "author": entry.find('span', {'class':'author'}).contents[1].string,
                "institution": 'Duke University',
                "year": entry.find('span', {'class':'date'}).string,
                "subject": subject,
                "url": 'http://dukespace.lib.duke.edu/' + entry.find('a')['href']
            }
        except Exception as e:
            print('skipping', entry)
            print('error', e)
            continue;
        print(dictionary)
        entries.append(dictionary)
    return entries

def get_page(URL):
    r = requests.get(URL)
    t = r.text
    return BeautifulSoup(t, 'html.parser')



mathSoup = []
mathSoup.append(get_page('http://dukespace.lib.duke.edu/dspace/handle/10161/1/browse?order=ASC&rpp=100&sort_by=1&value=Mathematics&etal=-1&offset=0&type=department'))
mathSoup.append(get_page('http://dukespace.lib.duke.edu/dspace/handle/10161/1/browse?order=ASC&rpp=100&sort_by=1&value=Mathematics&etal=-1&offset=100&type=department'))



csSoup = []
csSoup.append(get_page('http://dukespace.lib.duke.edu/dspace/handle/10161/1/browse?order=ASC&rpp=80&sort_by=1&value=Computer+Science&etal=-1&offset=80&type=department'))
csSoup.append(get_page('http://dukespace.lib.duke.edu/dspace/handle/10161/1/browse?order=ASC&rpp=80&sort_by=1&value=Computer+Science&etal=-1&offset=0&type=department'))

entries = []

for soup in mathSoup:
    entries = entries + run(soup, 'mathematics')
for soup in csSoup:
    entries = entries + run(soup, 'computer science')

with open("dukeDissertations.json", "w") as f:
    json.dump(entries, f, indent = 2)
```
