```python
import requests, json, sys, time, re
from bs4 import BeautifulSoup

def run(soup, subject):
    entries = []
    for entry in soup.findAll('li', {'class':'ds-artifact-item'}):
        #print(entry)
        print('thing', entry.find('span', {'class':'author'}).contents)
        #try:
        dictionary = {
            "title": entry.find('a').string,
            "author": entry.find('span', {'class':'author'}).contents[0].string,
            "institution": 'University of Maryland',
            "year": entry.find('span', {'class':'date'}).string,
            "subject": subject,
            "url": 'http://drum.lib.umd.edu/' + entry.find('a')['href']
        }
        #except Exception as e:
         #   print('skipping', entry)
          #  print('error', e)
           # continue;
        if(dictionary['author'][-1] == '-'): #Some authors have some date info. Trim it
            dictionary['author'] = dictionary['author'][:-7]
        print(dictionary)
        entries.append(dictionary)
    return entries

def get_page(URL):
    r = requests.get(URL)
    t = r.text
    return BeautifulSoup(t, 'html.parser')



mathSoup = []
mathSoup.append(get_page('http://drum.lib.umd.edu/handle/1903/2793/browse?order=ASC&rpp=400&sort_by=2&etal=-1&offset=0&type=dateissued'))



csSoup = []
csSoup.append(get_page('http://drum.lib.umd.edu/handle/1903/2756/browse?order=ASC&rpp=200&sort_by=2&etal=-1&offset=0&type=dateissued'))
csSoup.append(get_page('http://drum.lib.umd.edu/handle/1903/2756/browse?order=ASC&rpp=200&sort_by=2&etal=-1&offset=200&type=dateissued'))
csSoup.append(get_page('http://drum.lib.umd.edu/handle/1903/2756/browse?order=ASC&rpp=200&sort_by=2&etal=-1&offset=400&type=dateissued'))

entries = []

for soup in mathSoup:
    entries = entries + run(soup, 'mathematics')
for soup in csSoup:
    entries = entries + run(soup, 'computer science')

with open("marylandDissertations.json", "w") as f:
    json.dump(entries, f, indent = 2)
```
